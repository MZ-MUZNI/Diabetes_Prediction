# Part B – Compare and Evaluate AI Techniques

Application of Techniques in Diabetes Prediction

1. Support Vector Machines (SVM)

Support Vector Machines (SVM) is a powerful supervised learning technique that excels in classification tasks, especially with small-to-medium-sized datasets.

Applications in Diabetes Prediction:

  Binary Classification: SVM can classify patients as diabetic or non-diabetic based on features like glucose levels, BMI, and age.
  Non-linear Relationships: With the use of kernel functions (e.g., RBF kernel), SVM can capture complex non-linear relationships in the data, enhancing prediction accuracy.

Strengths:

  Effective in high-dimensional spaces, even when the number of dimensions exceeds the number of samples.
  Robust to overfitting, especially when a proper kernel is selected and regularization is applied.

Weaknesses:

  Computationally expensive for large datasets due to the quadratic time complexity of training.
  Requires careful selection of kernel functions and hyperparameters (e.g., C and gamma), which can be challenging.

Outputs:
The model outputs class labels (diabetic or non-diabetic) and margins that indicate confidence in predictions. Cross-validation ensures robust model performance.

2. Decision Trees

Decision Trees are intuitive machine learning models that use a tree-like structure to make decisions based on input features.

Applications in Diabetes Prediction:

  Predictive Modeling: Decision trees predict diabetes status by splitting the dataset based on key features like glucose levels, insulin, and BMI.
  Risk Assessment: The hierarchy of decisions helps identify high-risk individuals for diabetes.

Strengths:

  Easy to interpret and visualize, making it accessible for non-technical users in healthcare.
  Handles categorical and numerical data without requiring extensive preprocessing.

Weaknesses:

  Prone to overfitting, especially with deep trees, unless regularization techniques (e.g., pruning) are applied.
  Can be unstable to small changes in the dataset, leading to different tree structures.

Outputs:
The model outputs classifications (diabetic or non-diabetic) and feature importance scores, guiding healthcare interventions.

3. K-Nearest Neighbors (KNN)

KNN is a simple yet effective algorithm for classification problems, relying on the similarity of data points.

Applications in Diabetes Prediction:

  Patient Similarity: KNN identifies patients with similar medical profiles and assigns diabetes status based on their majority class.
  Anomaly Detection: By analyzing distances, KNN can flag anomalous cases for further investigation.

Strengths:

  Simple to implement and intuitive to understand, requiring no explicit training phase.
  Adapts well to non-linear decision boundaries, making it suitable for complex datasets.

Weaknesses:

  Computationally intensive during prediction, especially for large datasets, as it requires calculating distances for all data points.
  Sensitive to noisy data and outliers, which can adversely affect accuracy.

Outputs:
The algorithm outputs classifications and can visualize clusters of similar patients, aiding in personalized treatment.

```python
#Importing the libraries related to the project

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### a)Input Data

In here, I use Support Vector Machine(SVM) algorithm to implement this project. A set of labeled training data is necessary for Support Vector Machine (SVM) classifiers to learn from. A matrix with each row denoting a sample and each column denoting a feature should be used as the input data. The label or class of the sample should be in the final column. Integers, which are typically used to represent numbers, should be used as the class labels. Before using the input data to train the SVM classifier, preprocessing and normalization should be performed as needed.

The dataset used in this project got by Kaggle free downloadable. The dataset is preprocessed when it checked. All the data in this dataset is numeric and There are 9 attributes in dataset which are,

- 	Pregnancies – Number of times Pregnant
- 	Glucose - Plasma glucose concentration 2 hours in an oral glucose tolerance test
- 	BloodPressure – Diastolic blood pressure (mm Hg)
- 	SkinThickness – Triceps skin fold thickness (mm)
- 	Insulin – 2- Hour’s serum insulin (mu U/ml)
- 	BMI – Body Mass Index (weight in kg/height in m) ^ 2)
- 	DiabetesPedigreeFunction – Diabetes Pedigree Function
- 	Age – Age (Years)
- 	Outcome – Class Variable
    * 0 – non-Diabetic
    * 1 - Diabetic

```python
#seperating the data and labels
#X user droping a particular column name outcome in dataset
#if user droping a column axis must be equal to 1
#if user droping a rown axis must be equal to 0
#Y is getting only outcome dataset

X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
#user can get the all data except outcome
print(X)
```

```python
#user can get all the outcome results
print(Y)
```

### Data Standardization

standardize the data in a particular range that helps machine learning model to make better predictions

```python
#taking one instance of standard scaler function

scaler = StandardScaler()
```

```python
#fit the data into scaler

scaler.fit(X)
```

```python
#Fitting all inconsistent data with standard scaler function
standardized_data = scaler.transform(X)
```

```python
#getting the standardized data
print(standardized_data)
```

```python
#X represent the data and Y represents the model

X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
# Visualizing the input data(dataset)
standardized_df = pd.DataFrame(X, columns=diabetes_dataset.columns[:-1])
standardized_df['Outcome'] = Y
sns.pairplot(standardized_df, hue='Outcome', palette='husl', diag_kind='kde')
plt.suptitle('Pairplot of Standardized Features', y=1.02)
plt.show()
```

### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
#768 total participants in dataset
#614 participants going to use as training data
#154 participants use as test data

print(X.shape, X_train.shape, X_test.shape)
```

### Training The Model

```python
#SVC represents Support Vector Classifier
#in here user using linear model

classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine classifier
#Traing data is X_train
#Labels for the X_train data is Y_train

classifier.fit(X_train, Y_train)
```

# Part D – Testing

The holdout method was chosen from a variety of testing techniques, including K-fold cross validation, Bootstrap sampling, and others. The dataset is split into two training sets and one testing set as this method moves forward. The model accuracy is evaluated by comparing the expected results with the actual results on the testing set. Below is a summary of the training and testing set's accuracy levels

```python
#Accuracy score in the training data
#Predic the label for all X_train data and storing all these labels in X_train_prediction variable
#Comparing the prediction of model with the original labels that is Y_train

X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data: ', training_data_accuracy)
```

```python
#Accuracy score in the test data

X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data: ', test_data_accuracy)
```
